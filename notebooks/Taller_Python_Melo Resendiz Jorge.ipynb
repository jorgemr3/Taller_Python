```python
# Promedio de calificaciones (lista)
calif = [4, 6, 9, 7, 5]
suma = 0
for i in range(len(calif)):
    suma += calif[i]
print(suma / len(calif))
```

```python
# Transformación de lista: pares/impares
lista = [2, 456, 64, 234, 12, 45, 12, 45, 333, 332, 32, 12, 665, 43, 45, 22]
lista_2 = []
for i in range(len(lista)):
    if lista[i] % 2 == 0:
        lista_2.append(lista[i])
    else:
        lista[i] += 1
        lista_2.append(lista[i])
print(lista_2)
```

```python
# Manejo de datos con pandas: lectura segura del archivo de datos
import pandas as pd
import os
data_path = os.path.join('..', 'data', 'ejemplo.csv')
try:
    df = pd.read_csv(data_path)
    print(df.head())
    # Ejemplos de estadísticas
    if 'Parcial 1' in df.columns:
        print(df['Parcial 1'].mean())
        print(df['Parcial 1'].min())
        print(df['Parcial 1'].max())
except FileNotFoundError:
    print(f'Archivo no encontrado: {data_path}. Coloca tu archivo en data/ o ajusta la ruta.')
```

## Visualización con matplotlib

```python
# Gráfica de línea: temperatura vs tiempo (ejemplo)
import matplotlib.pyplot as plt
tiempo = [8, 10, 12, 13, 16, 18]
temperatura = [20, 23, 27, 25, 23, 19]
plt.plot(tiempo, temperatura)
plt.xlabel('Tiempo en horas')
plt.ylabel('Temperatura en Celsius')
plt.title('Temperatura vs Tiempo')
plt.grid(True)
plt.show()
```

```python
# Recuperar y graficar notas de un alumno si df está disponible
try:
    alumno = 5
    notas = df[df['ID'] == alumno][['Parcial 1', 'Parcial 2', 'Parcial 3']].iloc[0]
    plt.plot(['Parcial 1', 'Parcial 2', 'Parcial 3'], notas)
    plt.ylim(0, 10)
    plt.ylabel('Calificaciones')
    plt.title('Evolucion del alumno')
    plt.show()
except Exception as e:
    print('No se puede graficar la evolución del alumno:', e)
```

```python
# Añadir columna de promedio y estatus si df existe
import numpy as np
try:
    df['prom'] = df[['Parcial 1', 'Parcial 2', 'Parcial 3']].mean(axis=1)
    df['estatus'] = np.where(df['prom'] >= 7, 'Aprobado', 'Reprobado')
    print(df.head())
except Exception as e:
    print('No se pudo calcular promedios/estatus:', e)
```
